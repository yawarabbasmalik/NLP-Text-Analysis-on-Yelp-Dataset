```python
import nltk
import pandas as pd

# Read the dataset and extract the 'Review' column
df = pd.read_csv('Yelp Dataset2.csv')
```

```python
df.head()
```

```python
# Create a new DataFrame with 'Review' and 'Label' columns
df = df[['Review', 'Label']]

# Drop all other columns
df = df.drop(columns=[col for col in df.columns if col not in ['Review', 'Label']])
```

```python
df.head()
```

**Word Tokenization**

This code snippet applies tokenization on the "Review" column of the DataFrame df. The apply() method is used to iterate over each element in the "Review" column. The lambda function lambda x: word_tokenize(str(x)) is applied to each element, ensuring that the values are converted to strings using str() before tokenization.

```python
nltk.download('punkt')

from nltk.tokenize import word_tokenize


# Convert values in 'Review' column to string type
df['Review'] = df['Review'].astype(str)

# Apply tokenization on the 'Review' column
df['Tokenized_Review'] = df['Review'].apply(word_tokenize)
```

```python
print (df['Tokenized_Review'])
```

```python
df.head()
```

**Stop Words**

In this updated code, we first import the necessary libraries, including nltk and stopwords. We download the required resources using nltk.download() for both tokenization and stop words.

After tokenizing the reviews using word_tokenize(), we assign the tokenized results to the new column 'Tokenized_Review' in the DataFrame df.

Next, we proceed with stop words removal. We create a set of English stop words using stopwords.words('english'). Then, we apply a lambda function using apply() on the 'Tokenized_Review' column. The lambda function iterates over each review, removing any stop words using a list comprehension, and assigns the filtered results to the new column 'Filtered_Review' in the DataFrame df.

By incorporating these modifications, the code should now execute correctly, performing both tokenization and stop words removal on the 'Review' column of the DataFrame df.

```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords

# Stop words removal
stop_words = set(stopwords.words('english'))
df['Filtered_Review'] = df['Tokenized_Review'].apply(lambda x: [word for word in x if word.lower() not in stop_words])
```

```python
df.head()
```

**Lemmatization**

In this updated code, we first import the necessary libraries, including nltk, stopwords, and WordNetLemmatizer. We download the required resources using nltk.download() for tokenization, stopwords, and WordNet.

After tokenizing the reviews using word_tokenize(), we assign the tokenized results to the new column 'Tokenized_Review' in the DataFrame df.

Next, we proceed with stop words removal. We create a set of English stop words using stopwords.words('english'). Then, we apply a lambda function using apply() on the 'Tokenized_Review' column. The lambda function iterates over each review, removing any stop words using a list comprehension, and assigns the filtered results to the new column 'Filtered_Review' in the DataFrame df.

Finally, we perform lemmatization using the WordNetLemmatizer. We create an instance of the lemmatizer. Then, we apply a lambda function using apply() on the 'Filtered_Review' column. The lambda function iterates over each review, lemmatizing each word using the lemmatize() method, and assigns the lemmatized results to the new column 'Lemmatized_Review' in the DataFrame df.

By incorporating these modifications, the code should now correctly perform tokenization, stop words removal, and lemmatization on the 'Review' column of the DataFrame df.

```python
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer

# Lemmatization
lemmatizer = WordNetLemmatizer()
df['Lemmatized_Review'] = df['Filtered_Review'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
```

```python
df.head()
```

**Text Normalization**

```python
# Text normalization
df['Normalized_Review'] = df['Lemmatized_Review'].apply(lambda x: [word.lower() for word in x])
```

```python
df.head()
```

**Case Normalizataion**

In this adjusted code, we added the case normalization step after text normalization.

After performing text normalization, we assigned the normalized results to the new column 'Normalized_Review' in the DataFrame df.

Next, we proceeded with case normalization by applying two separate lambda functions. The first lambda function converts each word in the normalized review to lowercase using lower(), and the results are assigned to the new column 'Normalized_Review_Lower' in the DataFrame df. The second lambda function converts each word in the normalized review to uppercase using upper(), and the results are assigned to the new column 'Normalized_Review_Upper' in the DataFrame df.

By incorporating these modifications, the code will now perform tokenization, stop words removal, lemmatization, text normalization, and case normalization on the 'Review' column of the DataFrame df.

```python
# Case normalization
df['Normalized_Review_Lower'] = df['Normalized_Review'].apply(lambda x: [word.lower() for word in x])
df['Normalized_Review_Upper'] = df['Normalized_Review'].apply(lambda x: [word.upper() for word in x])
```

```python
df.head()
```

**Eliminating Null & Duplicate Values (Rows)**

In this adjusted code, we added the steps to remove null values and duplicates after the case normalization step.

After performing case normalization, we removed null values using the dropna() method. By specifying subset=['Normalized_Review_Lower'], we ensure that rows with null values in the 'Normalized_Review_Lower' column are dropped from the DataFrame df.

Next, we removed duplicates using the drop_duplicates() method. By specifying subset=['Normalized_Review_Lower'], we ensure that only rows with duplicated values in the 'Normalized_Review_Lower' column are dropped from the DataFrame df.

By incorporating these modifications, the code will now perform tokenization, stop words removal, lemmatization, text normalization, case normalization, removal of null values, and removal of duplicates on the 'Review' column of the DataFrame df.

```python
# Remove null values
df = df.dropna(subset=['Normalized_Review_Lower'])

# Remove duplicates
df = df.drop_duplicates(subset=['Normalized_Review_Lower'])
```

```python
df.head()
```

**Part-of-Speech (POS) tagging**

In this adjusted code, we added the step to perform Part-of-Speech (POS) tagging on the 'Normalized_Review_Lower' column of the DataFrame df.

After performing all the previous preprocessing steps, we apply the nltk.pos_tag function using apply() and assign the resulting POS tags to the new column 'POS_Tags' in the DataFrame df.

By incorporating these modifications, the code will now perform tokenization, stop words removal, lemmatization, text normalization, case normalization, removal of null values, removal of duplicates, and Part-of-Speech tagging on the 'Review' column of the DataFrame df.

```python
nltk.download('averaged_perceptron_tagger')

# Part-of-Speech tagging
df['POS_Tags'] = df['Normalized_Review_Lower'].apply(nltk.pos_tag)
```
